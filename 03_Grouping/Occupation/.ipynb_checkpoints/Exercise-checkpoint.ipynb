### Step 1. Import the necessary libraries

```python
import pandas as pd
import requests
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation").agg({'age': pd.Series.mean})
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

##### is the same as using the cross tab in the last question
##### this is more difficult approach but I want it to include them both
##### the easier approach is in the last Step

```python
# get the total number of people per occupation
all_per_occupation = users.groupby("occupation").agg({'gender': pd.Series.count})
all_per_occupation
```

```python
# get the number of males per occupation
males_per_occ = users[users.gender == "M"].groupby("occupation").agg({'gender': pd.Series.count})
males_per_occ.rename(columns={'gender': 'num_males'}, inplace=True)
males_per_occ
```

```python
# adding the males to the all_per_occupation
males_per_occ['total'] = all_per_occupation
males_per_occ
males_per_occ['males_percentage'] = (males_per_occ.num_males / males_per_occ.total) * 100

# sort the values from most to least
males_per_occ.sort_values(by="males_percentage", ascending=False, inplace=True)
males_per_occ.head()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation").age.describe()[["min", "max"]]
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation", "gender"]).agg({"age": pd.Series.mean})
```

### Step 8.  For each occupation present the percentage of women and men

```python
# easier approch
pd.crosstab(users.occupation, users.gender, normalize="index").round(4)*100
```
